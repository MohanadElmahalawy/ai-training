```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\Tareq\\Desktop\\HR_comma_sep.csv')

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar chart to show the impact of salaries on retention
plt.figure(figsize=(10, 10))
sns.countplot(x="salary", hue="left", data=df ,palette="Set2")
plt.title("Impact of Employee Salaries on Retention")
plt.xlabel("Salary Level")
plt.ylabel("Count")
plt.legend(title="Left", labels=["No", "Yes"])
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x="Department", hue="left", data=df ,palette="Set2")
plt.title("orelation between department and employee retention")
plt.xlabel("Department")
plt.ylabel("Count")
plt.legend(title="Left", labels=["No", "Yes"])
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x="satisfaction_level", hue="left", data=df ,palette="Set2")
plt.title("orelation between satisfaction and employee retention")
plt.xlabel("Satisfaction")
plt.ylabel("Count")
plt.legend(title="Left", labels=["No", "Yes"])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['satisfaction_level']],df.left,train_size=0.8)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_predicted=model.predict(X_test)
y_predicted
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test,y_test)
```
