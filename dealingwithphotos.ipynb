```python
import cv2 as cv
img=cv.imread('p.jpg')
img
```

```python
cv.imshow("new image",img)
cv.waitKey()
```

```python
blur=cv.GaussianBlur(img,(5,5),cv.BORDER_DEFAULT)
```

```python

cv.imshow("new image2",blur)
cv.waitKey(1)
```

```python
canny=cv.Canny(blur,115,77)
cv.imshow('canny edges',canny)
cv.waitKey(0)
```
