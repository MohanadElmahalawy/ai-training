```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

data = pd.read_csv("C:\\Users\\Tareq\\Downloads\\canada_per_capita_income.csv")
data
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(data['year'], data['per capita income (US$)'], color='red', marker='+', label='Data')
```

```python
year = data['year'].values.reshape(-1, 1)

print(year)
```

```python
income=data['per capita income (US$)']
print(income)
```

```python
reg=linear_model.LinearRegression()
reg.fit(year,income)
```

```python
reg.predict([[2020]])
```

```python
m=reg.coef_
print(m)
c=reg.intercept_
print(c)
#y=mx+c
y=m*2020+c
print(y)
```

```python
Predicted_income=reg.predict(year)
print(Predicted_income)
```

```python
Predicted_df=pd.DataFrame(year)
Predicted_df = Predicted_df.rename(columns={0: 'year'})
Predicted_df
```

```python
Predicted_df['predicted income'] = Predicted_income
Predicted_df
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(data['year'], data['per capita income (US$)'], color='red', marker='+', label='Data')
plt.plot(Predicted_df['year'],Predicted_df['predicted income'],color='orange')
```
